```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
from bs4 import BeautifulSoup
import time
import numpy as np
import csv
```

```python
def search_name(k):
    search_box = driver.find_element_by_class_name('input_search')
    search_box.clear()
    time.sleep(0.5)
    search_box.send_keys(k)
    time.sleep(0.5 + np.random.random())
    search_box.send_keys(Keys.ENTER)
    time.sleep(0.5 + np.random.random())
```

```python
driver = webdriver.Chrome('C:/Users/ehdtn/Desktop/Python/chromedriver')
driver.get('https://map.naver.com/')

driver.find_element_by_id('intro_popup_close').click()
time.sleep(0.5)

driver.find_element_by_class_name('span_search').click()
time.sleep(0.5)

df = pd.read_csv("C:/Users/ehdtn/Desktop/MN/0315.csv", engine="python")
```

```python
with open("C:/Users/ehdtn/Desktop/MN/0315_res.csv", "a", newline='') as f:
    wr = csv.writer(f, delimiter="|")
#     wr.writerow(["index", "keyword", "naver_nm", "naver_addr"])
    for i in range(88,150):
        if i % 50 == 0:
            time.sleep(7 + np.random.random()) 
        if i % 100 == 0:
            time.sleep(10 + np.random.random())
        n = df.iloc[i].NUM
        k = df.iloc[i].KEYWORD
        print (i)
        search_name(k)
        try:
            driver.find_element_by_class_name('notice_box')
        except:
            bs = BeautifulSoup(driver.page_source, 'html.parser')
            time.sleep(0.3)
            title = bs.find('span', class_='search_title_text').text
            address = bs.find('span', class_='search_text address').text
        else:
            title = ""
            address = ""
            time.sleep(0.3)
        finally:
            wr.writerow([n, k, title, address])
            
driver.quit()
```
